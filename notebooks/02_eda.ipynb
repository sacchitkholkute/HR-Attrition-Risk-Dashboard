# HR Attrition Analysis : Exploratory Data Analysis (EDA)

## Objective
Explore relationships between employee features and attrition to identify patterns and key drivers contributing to employee turnover.


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df=pd.read_csv("../data/raw/IBM_hr_attrition.csv")
df.head()
```

```python
plt.figure(figsize=(16,5))
sns.histplot(x=df["Attrition"],data=df)
plt.title("ATTRITION HISTOGRAM")
plt.show()
```

We can see clear imbalance in data , this validates our earlier observation from 01_data_understanding.ipynb

Due to the imabalance even if a false model predicts employees staying to be 100% he will achieve high accuracy 

```python
plt.figure(figsize=(16,5))
sns.boxplot(x=df['Attrition'],y=df['Age'])
plt.title("Age vs Attrition")
plt.show()
```

Younger employees show slightly higher attrition, though the difference is not very strong. Age alone may not be a decisive factor.

```python
df['OverTime']
```

for categorical vs categorical i will use countplot 
histplot for numerical distributions 


```python
plt.figure(figsize=(16,5))
sns.countplot(x="OverTime",hue="Attrition",data=df)
plt.title("OverTime vs Attrition")
plt.show()
```

Employees with overtime : Approx 150 Leave & approx 300 Stay
Employees without overtime : Approx 80-100 leave & Approx 1000-1100 stay

HR action i would suggest is to increase overtime pay or lessen the overtime hours to ensure employee wellness and avoiding burnout and job dissatisfaction in employees

```python
plt.figure(figsize=(16,5))
sns.boxplot(x="Attrition",y="MonthlyIncome",data=df)
plt.title("MonthlyIncome vs Attrition")
plt.show()
```

here it shows that employees having typically lower salary ranges have higher chance of leaving than the ones in the higher range . 
this shows that there is an employee dissatisfaction with lower salary range making them go for other jobs and explore other opportunities .
